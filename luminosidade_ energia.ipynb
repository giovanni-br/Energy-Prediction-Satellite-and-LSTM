```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense, Dropout
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import tensorflow as tf
#from datetime import datetime


df = sheet[['energy', 'lights']]

#Separate dates for future plotting

df_for_training = df


#LSTM uses sigmoid and tanh that are sensitive to magnitude so values need to be normalized
# normalize the dataset
scaler = StandardScaler()
scaler = scaler.fit(df_for_training)
df_for_training_scaled = scaler.transform(df_for_training)


#As required for LSTM networks, we require to reshape an input data into n_samples x timesteps x n_features. 
#In this example, the n_features is 2. We will make timesteps = 3. 
#With this, the resultant n_samples is 5 (as the input data has 9 rows).
trainX = []
trainY = []

n_future = 1   # Number of days we want to predict into the future
n_past = 3     # Number of past days we want to use to predict the future

for i in range(n_past, (int(0.75 *len(df_for_training_scaled))) - n_future +1):
    trainX.append(df_for_training_scaled[i - n_past:i, 0:df_for_training.shape[1]])
    trainY.append(df_for_training_scaled[i + n_future - 1:i + n_future, 0])

trainX, trainY = np.array(trainX), np.array(trainY)

print('trainX shape == {}.'.format(trainX.shape))
print('trainY shape == {}.'.format(trainY.shape))


# define Autoencoder model

model = Sequential()
model.add(LSTM(128, input_shape=(n_past, 2), return_sequences=True))
model.add(tf.keras.layers.LeakyReLU(alpha = 0.5))
model.add(LSTM(128, return_sequences=True))
model.add(tf.keras.layers.LeakyReLU(alpha = 0.5))
model.add(tf.keras.layers.Dropout(0.3))
model.add(LSTM(64, return_sequences=False))
model.add(tf.keras.layers.Dropout(0.3))
model.add(Dense(1))
#model = Sequential()
#model.add(LSTM(128, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
#model.add(LSTM(64, activation='relu', return_sequences=False))
#model.add(Dropout(0.2))
#model.add(Dense(trainY.shape[1]))

early_stopping = tf.keras.callbacks.EarlyStopping(monitor= 'val_loss', patience = 2, mode ='min')
model.compile(loss = tf.losses.MeanSquaredError(), optimizer= tf.optimizers.Adam(), metrics = [tf.metrics.MeanAbsoluteError()])

# fit model
history = model.fit(trainX, trainY, epochs=0, batch_size=72, validation_split = 0.1, verbose=1, shuffle = False)
model.summary()

#plt.plot(history.history['loss'], label='train')plt.plot(history.history['val_loss'], label='test')plt.legend()plt.show()

```
